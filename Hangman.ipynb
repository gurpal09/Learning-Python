```python
# Function to get the length of text file of words
def get_count():
    counter = 0
    with open('sowpods.txt','r') as f:
        line = f.readline()
        while line:
           # print(line)
            line = f.readline()
            counter +=1        
    return counter

# Function to return a random word
def pick_word(count):
    import random
    r = random.randint(0,count)

    counter = 0
    with open('sowpods.txt','r') as f:
        line = f.readline()
        while line:
            if counter == r:
                line = line[0:len(line)-1]
                return line.lower()
            line = f.readline()
            counter += 1

# Function to prompt user for input            
def get_input():
    while True:
        letter = input('Enter your letter: ')
    
        if letter.isalpha() and len(letter) == 1:
            break
    letter.lower()
    return letter

# Function to print word to screen
def print_to_screen(letter, word, myset):
    for letter in word: 
        if letter in myset:
            print('_ ',end='')
        else:
            print(letter + ' ',end ='')
            
def make_move(word):
    guessword = ''
    guesscounter = 0
    turn = 0
    # Keep track of used letters 
    used = set()

    winning_letters = set(word)

    while True:
    
        guess = get_input()
    
        # Case where same letter gets guessed twice
        while guess in used:
            print('This letter has already been guessed. Guess Again!')
            guess = get_input()
    
        # Uncomment this to see random word that is picked
        #print('Word: ',guessword)
    
        # Check if guess is in the word and remove letter from that set
        if guess in winning_letters:
            winning_letters.remove(guess)
          
        # if letter is not in word then penalize
        if guess not in word:
            guesscounter += 1
       
        # Function call to draw hangman
        draw_figure(guesscounter) 
        
        # Check for turn condition
        if guesscounter > 5:
            print('Game Over! You have used all guesses')
            print('The correct word was: ', word)
            return -1
        
        turn +=1
        used.add(guess)
    
        # Print to screen
        print_to_screen(guess, word, winning_letters)
    
        # Break the loop when the winning letter set is empty
        if len(winning_letters) == 0:
            print('Game Over. Winner!!!')
            break
        
    
    print('Game lasted '+ str(turn) + ' turns')

# Function to prompt user for new game or end session
def new_game(count):
    while True:
        if count == 0:
            print('********************Welcome to Hangman********************')
            count = get_count()
            word = pick_word(count)
            
            # Uncomment for debug
            # print('The word is: ',word)
            make_move(word)
        else:
            usr_input = input('Enter y to play another game or n to exit: ')
            if usr_input == 'y':
                print('Starting new game...')
                count = get_count()
                word = pick_word(count)
                # print('The word is: ',word)
                print('********************Welcome to Hangman********************')
                make_move(word)
            elif usr_input == 'n':
                print('Ending Session. Thanks for playing...')
                break
                
def draw_figure(count):
    if count == 0:
        return
    elif count == 1:
        print('|||||||||||')
        print('|  o   o  |')
        print(' |   __  |')
        print('  \  __ /')
    elif count == 2:
        print(' |||||||||||')
        print(' |  o   o  |')
        print('  |   __  |')
        print('   \  __ /')
        print('____|   |____')
        print('|            |')
        print('|    GUCCI   |')
        print('|            |')
        print('\            /')
        print(' \__________/')
    elif count == 3:
        print(' |||||||||||')
        print(' |  o   o  |')
        print('  |   __  |')
        print('   \  __ /')
        print('____|   |________')
        print('|            | | |')
        print('|    GUCCI   | | |')
        print('|            | | |')
        print('\            / | |')
        print(' \__________/  |_|')
    elif count == 4:
        print('    |||||||||||')
        print('    |  o   o  |')
        print('     |   __  |')
        print('      \  __ /')
        print('_______|   |________')
        print('| |             | |')
        print('| |    GUCCI    | |')
        print('| |             | |')
        print('| |\           /| |')
        print('|_| \_________/ |_|')
    elif count == 5:
        print('    |||||||||||')
        print('    |  o   o  |')
        print('     |   __  |')
        print('      \  __ /')
        print('_______|   |________')
        print('| |             | |')
        print('| |    GUCCI    | |')
        print('| |             | |')
        print('| |\           /| |')
        print('|_| \_________/ |_|')
        print('     |  |')
        print('     |  |')
        print('     |  |')
        print('     |  |')
        print('   |____|')
    elif count == 6:
        print('    GAME OVER!!!')
        print('    |||||||||||')
        print('    |  o   o  |')
        print('     |   __  |')
        print('      \  __ /')
        print('_______|   |_______')
        print('| |             | |')
        print('| |    GUCCI    | |')
        print('| |             | |')
        print('| |\           /| |')
        print('|_| \_________/ |_|')
        print('     |  |  |  |')
        print('     |  |  |  |')
        print('     |  |  |  |')
        print('     |  |  |  |')
        print('   |____|  |____|')
        
    print(str(6-count), ' Guesses Remaining.')
```
