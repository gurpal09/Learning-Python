Given two .txt files that have lists of numbers in them, find the numbers that are overlapping. One .txt file has a list of all prime numbers under 1000, and the other .txt file has a list of happy numbers up to 1000.

(If you forgot, prime numbers are numbers that canâ€™t be divided by any other number. And yes, happy numbers are a real thing in mathematics - you can look it up on Wikipedia.

```python
# Reading Prime Numbers data
primenums = set()

with open('primenumbers.txt','r') as open_file:
    line = open_file.readline()
    while line:
        # print(line)
        line = line.replace("\n","")
        primenums.add(line)
        line = open_file.readline()
        
```

```python
# Now reading happynumbers file

happynums = set()

with open('happynumbers.txt','r') as open_file:
    line = open_file.readline()
    while line:
        # print(line)
        line = line.replace("\n","")
        happynums.add(line)
        line = open_file.readline()
```

```python
print(happynums)
print(primenums)
```

```python
# We can compare using the union function for sets

match = primenums.union(happynums)
for n in match:
    print(n)
```
